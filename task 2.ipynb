```python
preferred_cuisines = ['Turkish', 'Seafood, Asian, Filipino, Indian']  # Multiple cuisines
preferred_cities = ['IIstanbul', 'Mandaluyong City']  # Multiple cities
preferred_price_range = 3
min_rating = 4.0
```

```python
def recommend_restaurants(df, cuisine_preferences, price_range, min_rating, cities):
    # Strip any extra whitespace
    df['City'] = df['City'].str.strip()
    df['Cuisines'] = df['Cuisines'].str.strip()
    
    # Filter based on city, price range, and rating
    filtered_df = df[
        (df['City'].isin(cities)) & 
        (df['Price range'] == price_range) &
        (df['Aggregate rating'] >= min_rating)
    ]
    if cuisine_preferences:
        # Create a regex pattern for multiple cuisines
        pattern = '|'.join(cuisine_preferences)
        filtered_df = filtered_df[filtered_df['Cuisines'].str.contains(pattern, case=False, na=False)]
    
    return filtered_df
    
```

```python
recommended_restaurants = recommend_restaurants(df, preferred_cuisines, preferred_price_range, min_rating, preferred_cities)
recommended_restaurants


```
